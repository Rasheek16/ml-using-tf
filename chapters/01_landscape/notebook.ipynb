```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
from sklearn.linear_model import LinearRegression
```

```python
data_root = "https://raw.githubusercontent.com/ageron/data/refs/heads/main/lifesat/lifesat.csv"
lifesat = pd.read_csv(data_root)
X=lifesat[["GDP per capita (USD)"]].values
y=lifesat["Life satisfaction"].values
```

```python
lifesat.plot.scatter(x="GDP per capita (USD)", y="Life satisfaction", figsize=(10, 6), title="Life Satisfaction vs GDP per Capita") 
plt.axis([23_500,62_500, 4, 9])
plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
X_new = np.array([[37_655.2]])
```

```python
print(model.predict(X_new))  # Predict life satisfaction for GDP per capita of 37,655.2 USD
```
